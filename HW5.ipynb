### Exercise 2: Undersampling
Loading the Academy Building image should not be a problem. However, some images are too big to be processed easily on your computer. When an image is big, the resolution can be adjusted by undersampling your image. Undersampling means that instead of working with the full image, we only use every N'th column and row as our image. This can preserve the original image, while lowering the memory required to store the image. In this excercise we will undersample the 'Academy_building.jpg' image by a factor of N (N'th pixel!) in both the x and y directions after having loaded the entire image. Make sure that you start at index `0` and include all N'th pixels (including the last pixel if it it divisible by N).

```python
#undersample
# in: numpy array with 3 dimensions: image
#     integer containing the undersampling factor: N
# out: undersampled numpy array with 3 dimensions

def undersample(image, N):
    #Complete your function here
    NN = image[::N,::N].copy() # explicit copy, O(n)
    return NN
```

```python
#Example of a correctly formatted function output (actual output from dummy image)
print(np.array([[[0, 1, 2], [6, 7, 8], [12, 13, 14]], [[30, 31, 32], [36, 37, 38], [42, 43, 44]], [[60, 61, 62], [66, 67, 68], [72, 73, 74]]]))

dummy_image = np.arange(0, 75, 1)
dummy_image.shape = (5, 5, 3)

#Your output from dummy image
print(undersample(dummy_image, 2))
```

```python
#grayscale
# in: a three dimensional numpy array: image
# out: a two dimensional numpy array containing the mean over the third dimension (or grayscale image)

def grayscale(image):
    return image.mean(axis=2)
```

### Excercise 4: Pixel Value Distribution
Next we will take a look at the distribution of the pixel values in this image. For this we will again be using the grayscale version of 'Academy_building.jpg'. To explore which values the pixels have, you can plot a histogram (you can try this yourselves). This histogram would show you a sort of bimodal distribution, with a steep normal curve above 175 and a more flat normal curve below 175.

We would like to summarize the upper normal curve (of all pixel values above 175) by computing the mean and the standard deviation. In order to select only the pixels with a value above 175 you can use a mask. A mask is a very convenient object that returns numpy array values based on a condition set by the user. For example, if we want to return all even values of an array the syntax could look as follows:

`even = array[array%2 == 0]`

```python
#summary_pixels
# in: two dimensional numpy array: grayscale_image
# out: tuple containing the mean and standard deviation of the pixel distribution over 175
def summary_pixels(grayscale_image):
    #Complete your function here
    array = grayscale_image[grayscale_image > 175]
    mean = np.mean(array)
    std = np.std(array)
    return (mean,std)
```
