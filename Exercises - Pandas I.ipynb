## Extra Exercises

1. Find the number of new cases per country (select two columns: `New_cases`, `Country`) for a date, which was four days ago starting from today. Pefrom the task programically - do not put the actual date, use the package `datetime`.  

```python
from datetime import date, timedelta
 
today_date = date.today()
td = timedelta(-4)
daysago = today_date+td
newdata = covid19[covid19['Date_reported'] == str(daysago)][['Country','New_cases']]
newdata.head()
```
